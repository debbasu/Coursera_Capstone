import the necessary packages and down load the wiki page as a html file

then get the first table and parse the texts to get rid of newline characters

```python
!wget -q -O 'canada.html' https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M
with open('canada.html') as html_file:
    soup = BeautifulSoup(html_file, 'lxml')
table = soup.findAll('table', attrs={"class" : "wikitable sortable"})[0]

#print(table)
temphead = table.findAll('th')
headers = [header.text.replace('\n','') for header in temphead]
#print(headers)
rows = []

for row in table.findAll('tr'):
    rows.append([val.text.replace('\n','') for val in row.find_all('td')])
#print(rows)
with open('output_file.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(headers)
    writer.writerows(row for row in rows if row)

df = pd.read_csv('output_file.csv')
df = df[df.Borough != 'Not assigned']
df.Neighbourhood[df.Neighbourhood == 'Not assigned'] = df.Borough
df.dtypes
df = df.groupby('Postcode').agg({
                             'Neighbourhood': ', '.join, 
                             'Borough':'first' }).reset_index()
df.head(1500)
```

```python
df.shape 
```

```python
!wget -q -O 'coordinate.csv' http://cocl.us/Geospatial_data
df = df.rename(columns={'Postcode': 'PostalCode'})
df_coordinate = pd.read_csv('coordinate.csv')
df_coordinate = df_coordinate.rename(columns={'Postal Code': 'PostalCode'})
df_coordinate.dtypes
```

```python
result = pd.merge(df,
                 df_coordinate,
                 on='PostalCode')
```

```python
result.head(1500)
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for Latitude, Longitude, Borough, Neighbourhood in zip(result['Latitude'], result['Longitude'], result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [Latitude, Longitude],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
